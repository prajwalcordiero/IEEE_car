```python
import joblib
import pandas as pd

# 1. Load the model you downloaded
model_path = 'accident_model.pkl'
try:
    model = joblib.load(model_path)
    print("✅ Model loaded successfully!")
except FileNotFoundError:
    print(f"❌ Error: {model_path} not found in this folder.")
    exit()

current_driving_data = {
    'Weather': 1,              
    'Road_Type': 0,            
    'Time_of_Day': 2,         
    'Traffic_Density': 1,      
    'Speed_Limit': 80,         
    'Number_of_Vehicles': 3,   
    'Driver_Alcohol': 0,       
    'Road_Condition': 1,      
    'Vehicle_Type': 0,         
    'Driver_Age': 25,          
    'Driver_Experience': 5,    
    'Road_Light_Condition': 1  
}

# 3. Convert to a DataFrame (the model needs names to avoid errors)
input_df = pd.DataFrame([current_driving_data])

# 4. Predict
prediction = model.predict(input_df)[0]
probability = model.predict_proba(input_df)[0][1]

# 5. Output result
print("-" * 30)
if prediction == 1:
    print(f"🚨 WARNING: High Accident Risk Detected!")
    print(f"Confidence Level: {probability*100:.2f}%")
else:
    print(f"✅ STATUS: Safe Driving Environment.")
    print(f"Risk Level: {probability*100:.2f}%")
print("-" * 30)
```
